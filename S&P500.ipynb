```python
# Ordenar el DataFrame por símbolo y fecha
merged_df = merged_df.sort_values(by=['Symbol', 'Date'])

# Calcular la SMA de 50 días para cada símbolo
merged_df['SMA_50'] = merged_df.groupby('Symbol')['Close'].rolling(window=50).mean().reset_index(drop=True)

# Calcular la SMA de 200 días para cada símbolo
merged_df['SMA_200'] = merged_df.groupby('Symbol')['Close'].rolling(window=200).mean().reset_index(drop=True)

# Calcular la EMA de 12 días para cada símbolo
merged_df['EMA_12'] = merged_df.groupby('Symbol')['Close'].apply(lambda x: x.ewm(span=12).mean()).reset_index(drop=True)

# Calcular la EMA de 26 días para cada símbolo
merged_df['EMA_26'] = merged_df.groupby('Symbol')['Close'].apply(lambda x: x.ewm(span=26).mean()).reset_index(drop=True)
```

**CRUCE DE MEDIAS SIMPLES SMA** Cuando la SMA de 50 días cruza por encima de la SMA de 200 días, se considera una señal de compra (cruce dorado), mientras que cuando la SMA de 50 días cruza por debajo de la SMA de 200 días, se considera una señal de venta (cruce de la muerte).

```python
import plotly.graph_objects as go

symbol = 'MMM'  # Reemplaza 
# Filtrar el DataFrame para obtener solo los datos del símbolo específico
symbol_data = merged_df[merged_df['Symbol'] == symbol]

# Crear un gráfico de líneas con Plotly
fig = go.Figure()

# Agregar el precio de cierre al gráfico
fig.add_trace(go.Scatter(x=symbol_data['Date'], y=symbol_data['Close'],
                         mode='lines', name='Close'))

# Agregar la SMA de 50 días al gráfico
fig.add_trace(go.Scatter(x=symbol_data['Date'], y=symbol_data['SMA_50'],
                         mode='lines', name='SMA 50'))

# Agregar la SMA de 200 días al gráfico
fig.add_trace(go.Scatter(x=symbol_data['Date'], y=symbol_data['SMA_200'],
                         mode='lines', name='SMA 200'))

# Personalizar el gráfico (título, ejes, etc.)
fig.update_layout(title=f'Cruce de medias móviles para {symbol}',
                  xaxis_title='Fecha',
                  yaxis_title='Precio')

# Mostrar el gráfico
fig.show()
```

```python
# Calcula los rendimientos diarios
merged_df['returns'] = merged_df.groupby('Symbol')['Close'].pct_change()

# Calcula el rendimiento medio y la desviación estándar de los rendimientos diarios
stats = merged_df.groupby('Symbol')['returns'].agg(['mean', 'std']).reset_index()

# Define el rendimiento libre de riesgo
risk_free_rate = 0.02

# Calcula el ratio de Sharpe
stats['sharpe_ratio'] = (stats['mean'] - risk_free_rate) / stats['std']

# Crea el gráfico de dispersión
plt.scatter(stats['mean'], stats['std'], c=stats['sharpe_ratio'], cmap='viridis', marker='o')
plt.colorbar(label='Sharpe Ratio')

# Agrega las líneas de cuadrantes
plt.axvline(stats['mean'].mean(), color='r', linestyle='--')
plt.axhline(stats['std'].mean(), color='r', linestyle='--')

# Etiquetas de los ejes y título
plt
```

```python
# Define el tamaño del gráfico
plt.figure(figsize=(20, 15))

# Crea el gráfico de dispersión
scatter = plt.scatter(stats['mean'], stats['std'], c=stats['sharpe_ratio'], cmap='viridis', marker='o')
plt.colorbar(label='Sharpe Ratio')

# Agrega las líneas de cuadrantes
plt.axvline(stats['mean'].mean(), color='r', linestyle='--')
plt.axhline(stats['std'].mean(), color='r', linestyle='--')

# Etiquetas de los ejes y título
plt.xlabel('Rendimiento medio')
plt.ylabel('Desviación estándar')
plt.title('Ratios de Sharpe con cuadrantes')

# Agrega el símbolo de la empresa a cada punto en el gráfico
for i, txt in enumerate(stats['Symbol']):
    plt.annotate(txt, (stats['mean'][i], stats['std'][i]), fontsize=8, xytext=(5, 0), textcoords='offset points')

# Muestra el gráfico
plt.show()
```

```python
# Determina el punto de corte para los cuadrantes
mean_cutoff = stats['mean'].mean()
std_cutoff = stats['std'].mean()

# Asigna cada empresa a un cuadrante
def assign_quadrant(row):
    if row['mean'] >= mean_cutoff and row['std'] >= std_cutoff:
        return 'Q1'
    elif row['mean'] < mean_cutoff and row['std'] >= std_cutoff:
        return 'Q2'
    elif row['mean'] < mean_cutoff and row['std'] < std_cutoff:
        return 'Q3'
    else:
        return 'Q4'

stats['quadrant'] = stats.apply(assign_quadrant, axis=1)

# Calcula el rendimiento medio por cuadrante
quadrant_returns = stats.groupby('quadrant')['mean'].mean()

# Imprime los resultados
print("Rendimiento medio por cuadrante:")
print(quadrant_returns)
```
